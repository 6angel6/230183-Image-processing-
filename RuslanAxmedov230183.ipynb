```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("image.jpeg")
image_RGB  = cv2.cvtcolor(image, cv2.COLOR_BGR2RGB)

sub_image = cv2.subtract(image_RGB, 100)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sub_image)
plt.title('sub_image')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("image.jpeg")
image_RGB  = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

added_image = cv2.add(image_RGB, 100)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original image')
plt.axis('off')

plt.subplot(1,2,1)
plt.imshow(added_image)
plt.title('Added image')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("image.jpeg")
image_RGB  = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

mul_image = cv2.multiply(image_RGB, 2)

plt.subplot((1,2,1)
plt.imshow(image_RGB)
plt.title('Original image')
plt.axis('off')

plt.subplot(1,2,1)
plt.imshow(mul_image)
plt.title('multiply image 2')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("image.jpeg")
image_RGB  = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

mul_image = cv2.divide(image_RGB, 2)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(mul_image)
plt.title('divide image 2')
plt.axis('off')
plt.show()
```
